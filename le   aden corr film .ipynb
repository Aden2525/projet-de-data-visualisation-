```python
## l' importattion de  bibilotheque  pour manipule et viasualiser le donnee 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
##   on va charger le donnee 
df=pd.read_csv('movies.csv')
```

```python
## verification global de donnee 
df.head()
```

```python
## le dimension de notre  jeux de donnee , dans notre jeux de donnee il-y-a 7668 ligne et 15 colonne qui decrire ce ligne 
df.shape
```

```python
## le type de donnee 
df.dtypes
```

```python
##verifications de ligne doublons , il-y-a pas de  ligne doublons dans le jeux de donnee 
df.duplicated().sum()
sns.scatterplot(x=df['budget'] , y=df['gross'])
```

```python
## verification de valeur manquante dans notre jeux de donnee 
for  i in df.columns :
    pct_missing=np.mean(df[i].isnull())
    print(f'{i} {pct_missing} %')
```

```python
## supprimer le valeur manquante 
df.dropna(how='any', inplace=True)
```

```python
##  convertire le type de donnee 

df['budget']=df['budget'].astype(int)

```

```python

pd.set_option('display.max_rows' , None )
```

```python
df.sort_values(by=['gross'] , inplace=False , ascending=True)
df.head()
```

```python
plt.scatter(x=df['budget'] , y=df['gross'])
```

```python
## verification de correlation entre le numerique
df.corr(method='pearson')
```

```python
## visualisation de correlaction entre variable numerique 
correlation_matrice=df.corr(method='pearson')
plt.figure(figsize=(5, 3))
sns.heatmap(correlation_matrice, annot=True)
plt.title('le caracteristique de film  ')
plt.xlabel('le caracteristique de film ')
```

```python
## encodage de variable qualitative pour visualiser le correlation de toute le jeux de donnee 
df_numeric=df
for df_colname in df.columns :
    if(df_numeric[df_colname].dtype=='object') :
        df_numeric[df_colname]=df_numeric[df_colname].astype('category')
        df_numeric[df_colname]=df_numeric[df_colname].cat.codes
df_numeric.head()    
```

```python
df_numeric.corr(method='pearson')
```

```python
correlation=df.corr(method='pearson')
plt.figure(figsize=(10,10))
sns.heatmap(correlation, annot=True)
plt.title('le caracteristique de film  ')
plt.xlabel('le caracteristique de film ')
```

```python
correlation_matrice= df_numeric.corr()
corr_paire=correlation_matrice.unstack()
corr_paire
```

```python
sorted_paire= corr_paire.sort_values(ascending=True)
sorted_paire
```

```python
high_corr=sorted_paire[(sorted_paire)>0.5]
high_corr
```
